```python
# KoboToolbox Data Analysis Project: Household Needs Assessment

# This notebook performs data loading, cleaning, descriptive statistics, and
# advanced visualizations for your KoboToolbox export data.
# It has been specifically updated to use the EXACT COLUMN HEADERS you provided.

# --- Part 1: Setup and Data Loading from Google Drive ---

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os # For managing file paths
from google.colab import drive

# Set plotting style for better aesthetics
sns.set_theme(style="whitegrid")

print("--- Part 1: Mounting Google Drive and Loading Your Data ---")
print("This step connects your Colab notebook to your Google Drive and loads your KoboToolbox CSV export.")

drive.mount('/content/drive')

# Define the full path to YOUR CSV file in Google Drive
# IMPORTANT:
# 1. Ensure the folder '/content/drive/My Drive/Colab_Kobo_Data/' exists in your Google Drive,
#    OR update 'drive_folder_path' to the actual folder where your CSV is located.
drive_folder_path = '/content/drive/My Drive/Portfolio/Data_analysis_projects/Assessing_the_Needs_of_Vulnerable_Households_for_an_NGO_in_Lebanon/_data/'
your_csv_file_name = 'cleaned_fake_form_entries.csv' # CONFIRMED: This is the file you provided!
csv_full_path = os.path.join(drive_folder_path, your_csv_file_name)

# Load the data from Google Drive into a DataFrame 'df'
print(f"Attempting to load data from: {csv_full_path}")
try:
    # Most Kobo exports are comma-delimited and UTF-8 encoded
    df = pd.read_csv(csv_full_path, on_bad_lines='warn')
    print("Data loaded successfully from Google Drive!")
    print("\nFirst 5 rows of your DataFrame (df):")
    print(df.head())
    print("\nDataFrame Info (shows columns, non-null counts, and data types):")
    df.info()
except FileNotFoundError:
    print(f"ERROR: File not found at {csv_full_path}. Please ensure:")
    print("  1. Your Google Drive is mounted correctly.")
    print("  2. The file name '{your_csv_file_name}' is exactly correct (including case and extension).")
    print("  3. The path '{drive_folder_path}' leads to the correct folder in your Drive.")
    print("  4. You have placed your CSV file in that location.")
    raise # Stop execution if file isn't found
except Exception as e:
    print(f"AN UNEXPECTED ERROR OCCURRED while reading the CSV: {e}")
    print("Possible issues:")
    print("  - Incorrect delimiter (e.g., file uses semicolon instead of comma). Try: `df = pd.read_csv(csv_full_path, delimiter=';')`")
    print("  - Malformed CSV, or extra lines before headers. Check your CSV file's content.")
    print("  - Incorrect encoding. Try: `df = pd.read_csv(csv_full_path, encoding='latin1')`")
    raise # Stop execution on other errors


# --- Part 2: Data Cleaning and Preparation for Analysis ---

print("\n--- Part 2: Data Cleaning and Preparation ---")
print("Filtering for 'complete' interviews and ensuring correct data types for analysis.")

# --- MAPPING OLD COLUMN NAMES USED IN SCRIPT TO YOUR NEW COLUMN NAMES ---
# This dictionary helps map the names used in the analysis logic to your actual CSV headers.
# This makes the script more robust to header changes.
COLUMN_MAP = {
    'interview_status': 'Interview Status',
    'household_size': 'Household Size',
    'disabilities': 'Disabilities in Household?',
    'disab': 'Type(s) of Disability',
    'disab_other': 'Other Disability Type',
    'head_household_legal_status': 'Head of Household Legal Status',
    'education_level': 'Head of Household Education Level',
    'education_level_other': 'Other Education Level',
    'employment_status': 'Head of Household Employment Status',
    'employment_status_other': 'Other Employment Status',
    'no_members_per_day': 'Number of Meals Per Day',
    'governorate': 'Governorate',
    'district': 'District',
    'consent_obtained': 'Consent Obtained',
    # For repeat group members, we'll generate these dynamically based on your pattern
    # 'member_X_age': 'Member X Age'
    # 'member_X_gender': 'Member X Gender'
}

# Rename columns in the DataFrame for easier use in the script
# Only rename if the original column exists in df.columns
df = df.rename(columns={v: k for k, v in COLUMN_MAP.items() if v in df.columns})

# Filter for 'complete' interviews for most analyses
# It's good practice to analyze refusal rates (or incomplete interviews) separately if needed.
# This ensures that subsequent analyses are based on fully completed forms.
if 'interview_status' in df.columns:
    df_completed = df[df['interview_status'] == 'Completed'].copy()
    print(f"Analyzing {len(df_completed)} 'complete' interviews out of {len(df)} total submissions.")
else:
    print("Warning: 'interview_status' column (mapped from 'Interview Status') not found. Proceeding with all data for analysis.")
    df_completed = df.copy() # Use all data if status column is missing


# Ensure relevant columns have appropriate data types
# Convert 'Member X Age' columns to numeric (NaN for empty strings/non-numeric)
# We need to find columns that match the 'Member X Age' pattern
member_age_cols_raw = [col for col in df_completed.columns if col.startswith('Member ') and ' Age' in col]
for col in member_age_cols_raw:
    df_completed[col] = pd.to_numeric(df_completed[col], errors='coerce')

# Convert other expected numerical survey responses to numbers, handling potential empty strings from skip logic
numeric_cols = ['household_size', 'no_members_per_day'] # These are now mapped names
for col in numeric_cols:
    if col in df_completed.columns:
        df_completed[col] = pd.to_numeric(df_completed[col], errors='coerce')

# Convert common categorical columns to 'category' type for efficiency and plotting
categorical_cols = [
    'governorate', 'district', 'interview_status', 'consent_obtained',
    'disabilities', 'head_household_legal_status', 'education_level',
    'employment_status'
]
for col in categorical_cols:
    if col in df_completed.columns:
        df_completed[col] = df_completed[col].astype('category')

print("\nData cleaning and type conversion complete. Your DataFrame 'df_completed' is ready for analysis.")


# --- Part 3: Descriptive Statistics for Key Questions ---

print("\n--- Part 3: Descriptive Statistics ---")
print("Answering basic summary questions about the data collected from completed interviews.")

# 1. What is the current status of the interview? (completed, refused, incomplete)
# Note: This analysis uses the *original* df to show all statuses.
if 'interview_status' in df.columns:
    print("\n## 1. Current Status of Interviews (All Submissions):")
    interview_status_counts = df['interview_status'].value_counts(dropna=False)
    print(interview_status_counts)
    plt.figure(figsize=(8, 5))
    sns.barplot(x=interview_status_counts.index, y=interview_status_counts.values, palette='viridis')
    plt.title('Distribution of Interview Status (All Submissions)')
    plt.xlabel('Interview Status')
    plt.ylabel('Number of Submissions')
    plt.show()
    print("Interpretation: This shows the breakdown of how many forms were completed, refused, or left incomplete, indicating data collection progress and potential refusal rates from the overall submissions.")
else:
    print("\n## 1. Current Status of Interviews: 'interview_status' column not found. Please check your raw CSV header for 'Interview Status'.")


# 2. What is the distribution of males and females in the household?
# Find 'Member X Gender' columns dynamically
# --- Corrected Code for Question 2 ---
# (Ensure you've run all cells before this, especially Part 1 and Part 2)

print("\n## 2. Distribution of Males and Females in Households (Completed Interviews):")

# --- DEBUGGING STEPS: Add these to diagnose the empty data issue ---
print("\n--- DEBUG INFO FOR GENDER DISTRIBUTION ---")
print("df_completed head (first 5 rows):")
print(df_completed.head()) # See what's actually in df_completed

print("\nIdentified raw gender columns:")
# This line dynamically finds columns like 'Member 1 Gender', 'Member 2 Gender' etc.
member_gender_cols_raw = [col for col in df_completed.columns if col.startswith('Member ') and ' Gender' in col]
print(member_gender_cols_raw)

if not member_gender_cols_raw:
    print("WARNING: No columns matching 'Member X Gender' pattern were found in df_completed.")
    print("Please check your original CSV headers for Member 1 Gender, Member 2 Gender, etc.")
else:
    print("\nNull counts for identified gender columns:")
    # Show how many missing values are in each identified gender column
    print(df_completed[member_gender_cols_raw].isnull().sum())
    print("\nNon-empty values in identified gender columns (sum of value_counts):")
    # Check if there's any actual data in these columns
    for col in member_gender_cols_raw:
        if col in df_completed.columns:
            non_null_count = df_completed[col].dropna().count()
            print(f"- {col}: {non_null_count} non-null values")
            if non_null_count > 0:
                print(df_completed[col].value_counts()) # Show values if present
    if df_completed[member_gender_cols_raw].stack().dropna().empty:
        print("\nCONCLUSION: Identified gender columns are either missing or entirely empty/NaN after stacking.")

print("--- END DEBUG INFO ---")


# --- ORIGINAL ANALYSIS CODE (with plt.pie fix) ---
if member_gender_cols_raw: # Proceed only if gender columns were found
    # Stack the gender columns into a single series, drop NaNs (from skipped members)
    all_member_genders = df_completed[member_gender_cols_raw].stack().reset_index(drop=True)
    sex_distribution = all_member_genders.value_counts()

    if not sex_distribution.empty: # Check if there's data after value_counts
        print(sex_distribution)
        plt.figure(figsize=(7, 5))
        # FIX: Changed sns.pie to plt.pie
        plt.pie(sex_distribution, labels=sex_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
        plt.title('Gender Distribution of Household Members')
        plt.ylabel('')
        plt.show()
        print("Interpretation: This provides the overall proportion of males and females across all individuals recorded in the completed households, giving a demographic overview of the surveyed population.")
    else:
        print("No valid gender data found in 'Member X Gender' columns to create distribution.")
else:
    print("Cannot perform gender distribution analysis as 'Member X Gender' columns were not found.")

# 3. What is the distribution of ages in the household?
# Find 'Member X Age' columns dynamically
member_age_cols_raw = [col for col in df_completed.columns if col.startswith('Member ') and ' Age' in col]
if member_age_cols_raw:
    print("\n## 3. Distribution of Ages in Households (Completed Interviews):")
    all_member_ages = df_completed[member_age_cols_raw].stack().reset_index(drop=True)
    all_member_ages = pd.to_numeric(all_member_ages, errors='coerce').dropna()

    print(all_member_ages.describe())
    plt.figure(figsize=(10, 6))
    sns.histplot(all_member_ages, bins=15, kde=True, color='skyblue')
    plt.title('Age Distribution of Household Members')
    plt.xlabel('Age (Years)')
    plt.ylabel('Number of Individuals')
    plt.show()
    print("Interpretation: The histogram visualizes the frequency of different age groups, while descriptive statistics summarize the central tendency and spread of ages across all individuals in completed households.")
else:
    print("\n## 3. Distribution of Ages: 'Member X Age' columns not found for analysis. Please check your raw CSV headers.")


# 4. Are there individuals with disabilities in the household? If yes, what kind of disabilities?
if 'disabilities' in df_completed.columns:
    print("\n## 4. Individuals with Disabilities (Completed Interviews):")
    disability_presence = df_completed['disabilities'].value_counts(dropna=False)
    print("Presence of individuals with disabilities in households:")
    print(disability_presence)

    plt.figure(figsize=(7, 5))
    sns.barplot(x=disability_presence.index, y=disability_presence.values, palette='coolwarm')
    plt.title('Presence of Individuals with Disabilities in Households')
    plt.xlabel('Disability Present?')
    plt.ylabel('Number of Households')
    plt.show()

    if 'disab' in df_completed.columns:
        df_disab_yes = df_completed[df_completed['disabilities'] == 'yes'].copy()
        if not df_disab_yes.empty:
            print("\nTypes of disabilities reported (from households with disabilities):")
            # The 'disab' column holds space-separated values for select_multiple
            all_disabilities = df_disab_yes['disab'].str.split(' ', expand=True).stack()
            disability_types_counts = all_disabilities.value_counts()
            print(disability_types_counts)

            plt.figure(figsize=(10, 6))
            sns.barplot(x=disability_types_counts.index, y=disability_types_counts.values, palette='magma')
            plt.title('Reported Disability Types')
            plt.xlabel('Disability Type')
            plt.ylabel('Number of Mentions')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()

            if 'disab_other' in df_disab_yes.columns:
                disab_other_count = df_disab_yes[df_disab_yes['disab_other'] != ''].shape[0]
                print(f"\nNumber of households with 'Other' disability specified: {disab_other_count}")
                if disab_other_count > 0:
                    print("Examples of 'Other' disability specifications:")
                    print(df_disab_yes[df_disab_yes['disab_other'] != '']['disab_other'].head())
        else:
            print("No 'yes' responses for disabilities in completed interviews to analyze disability types.")
    else:
        print("Warning: 'disab' column (mapped from 'Type(s) of Disability') not found to analyze disability types. Please check your raw CSV header.")
    print("Interpretation: Provides insights into the prevalence of disabilities across households and the specific types encountered, highlighting needs.")
else:
    print("\n## 4. Individuals with Disabilities: 'disabilities' column (mapped from 'Disabilities in Household?') not found.")


# 5. What is the legal status of the head of the household?
if 'head_household_legal_status' in df_completed.columns:
    print("\n## 5. Legal Status of Head of Household (Completed Interviews):")
    legal_status_counts = df_completed['head_household_legal_status'].value_counts(dropna=False)
    print(legal_status_counts)
    plt.figure(figsize=(8, 5))
    sns.barplot(x=legal_status_counts.index, y=legal_status_counts.values, palette='plasma')
    plt.title('Legal Status of Head of Household')
    plt.xlabel('Legal Status')
    plt.ylabel('Number of Households')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("Interpretation: Shows the distribution of different legal statuses (citizen, refugee, migrant) among household heads, crucial for understanding vulnerability and targeting assistance.")
else:
    print("\n## 5. Legal Status of Head of Household: 'head_household_legal_status' column (mapped from 'Head of Household Legal Status') not found.")


# 6. What is the highest education level of the head of the household?
if 'education_level' in df_completed.columns:
    print("\n## 6. Highest Education Level of Head of Household (Completed Interviews):")
    edu_level_counts = df_completed['education_level'].value_counts(dropna=False)
    print(edu_level_counts)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=edu_level_counts.index, y=edu_level_counts.values, palette='cubehelix')
    plt.title('Highest Education Level of Head of Household')
    plt.xlabel('Education Level')
    plt.ylabel('Number of Households')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    if 'education_level_other' in df_completed.columns:
        other_edu_count = df_completed[df_completed['education_level_other'] != ''].shape[0]
        print(f"\nNumber of households with 'Other' education specified: {other_edu_count}")
        if other_edu_count > 0:
            print("Examples of 'Other' education specifications:")
            print(df_completed[df_completed['education_level_other'] != '']['education_level_other'].head())
    print("Interpretation: Provides a snapshot of the educational attainment among household heads, which can correlate with economic opportunities and overall well-being.")
else:
    print("\n## 6. Highest Education Level: 'education_level' column (mapped from 'Head of Household Education Level') not found.")


# 7. What is the employment status of the head of the household?
if 'employment_status' in df_completed.columns:
    print("\n## 7. Employment Status of Head of Household (Completed Interviews):")
    emp_status_counts = df_completed['employment_status'].value_counts(dropna=False)
    print(emp_status_counts)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=emp_status_counts.index, y=emp_status_counts.values, palette='viridis')
    plt.title('Employment Status of Head of Household')
    plt.xlabel('Employment Status')
    plt.ylabel('Number of Households')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    if 'employment_status_other' in df_completed.columns:
        other_emp_count = df_completed[df_completed['employment_status_other'] != ''].shape[0]
        print(f"\nNumber of households with 'Other' employment specified: {other_emp_count}")
        if other_emp_count > 0:
            print("Examples of 'Other' employment specifications:")
            print(df_completed[df_completed['employment_status_other'] != '']['employment_status_other'].head())
    print("Interpretation: Reveals the economic activity and employment patterns of household heads, indicating levels of financial stability or precarity.")
else:
    print("\n## 7. Employment Status: 'employment_status' column (mapped from 'Head of Household Employment Status') not found.")


# 8. How many meals do the households consume per day?
if 'no_members_per_day' in df_completed.columns:
    print("\n## 8. Meals Consumed Per Day by Households (Completed Interviews):")
    meals_per_day_counts = df_completed['no_members_per_day'].value_counts(dropna=False).sort_index()
    print(meals_per_day_counts)
    plt.figure(figsize=(7, 5))
    sns.barplot(x=meals_per_day_counts.index, y=meals_per_day_counts.values, palette='Greens')
    plt.title('Number of Meals Consumed Per Day by Households')
    plt.xlabel('Meals Per Day')
    plt.ylabel('Number of Households')
    plt.show()
    print("Interpretation: This provides a direct measure of immediate food security status based on reported daily meal consumption.")
else:
    print("\n## 8. Meals Consumed Per Day: 'no_members_per_day' column (mapped from 'Number of Meals Per Day') not found.")


# --- Part 4: Advanced Visualizations and Relational Analysis ---

print("\n--- Part 4: Advanced Visualizations and Relational Analysis ---")
print("Exploring relationships between variables to gain deeper insights into the completed interviews.")

# 1. Visualizing data by Governorates (Proportions)
if 'governorate' in df_completed.columns:
    print("\n## 1. Visualizing data by Governorates (Proportions)")
    print("Objective: Understand how key indicators vary across different Governorates to identify regional disparities.")

    # Example 1: Proportion of Households with Disabilities by Governorate
    if 'disabilities' in df_completed.columns:
        plt.figure(figsize=(10, 6))
        disability_by_gov = df_completed.groupby('governorate')['disabilities'].value_counts(normalize=True).unstack().fillna(0)
        if 'yes' in disability_by_gov.columns: # Check if 'yes' is even a response
            sns.barplot(x=disability_by_gov.index, y=disability_by_gov['yes'], palette='viridis')
            plt.title('Proportion of Households with Disabilities by Governorate')
            plt.xlabel('Governorate')
            plt.ylabel('Proportion of Households with Disabilities (Yes)')
            plt.ylim(0, 1)
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()
            print("Interpretation: This bar chart shows the percentage of households in each governorate that reported having individuals with disabilities. Areas with higher bars indicate a higher prevalence, guiding regional resource allocation.")
        else:
            print("No 'yes' responses for disabilities found to visualize by governorate.")
    else:
        print("Warning: 'disabilities' column (mapped from 'Disabilities in Household?') not found for governorate analysis.")

    # Example 2: Average Meals Per Day by Governorate
    if 'no_members_per_day' in df_completed.columns:
        plt.figure(figsize=(10, 6))
        avg_meals_by_gov = df_completed.groupby('governorate')['no_members_per_day'].mean().sort_values(ascending=False)
        sns.barplot(x=avg_meals_by_gov.index, y=avg_meals_by_gov.values, palette='plasma')
        plt.title('Average Meals Consumed Per Day by Households per Governorate')
        plt.xlabel('Governorate')
        plt.ylabel('Average Meals Per Day')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
        print("Interpretation: This bar chart compares the average number of meals consumed daily across different governorates, potentially highlighting regions with higher food insecurity and identifying areas for focused intervention.")
    else:
        print("Warning: 'no_members_per_day' column (mapped from 'Number of Meals Per Day') not found for governorate analysis.")

    # Example 3: Stacked Bar Chart for Education Level by Governorate
    if 'education_level' in df_completed.columns:
        plt.figure(figsize=(12, 7))
        edu_by_gov = df_completed.groupby('governorate')['education_level'].value_counts(normalize=True).unstack().fillna(0)
        edu_by_gov.plot(kind='bar', stacked=True, figsize=(12,7), colormap='tab20')
        plt.title('Education Level Distribution by Governorate')
        plt.xlabel('Governorate')
        plt.ylabel('Proportion of Households')
        plt.xticks(rotation=45, ha='right')
        plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()
        print("Interpretation: A stacked bar chart like this shows the internal distribution of education levels within each governorate, allowing for comparison of educational attainment patterns and resource allocation needs across regions.")
    else:
        print("Warning: 'education_level' column (mapped from 'Head of Household Education Level') not found for governorate analysis.")
else:
    print("\n## 1. Visualizing data by Governorates: 'governorate' column not found. Please check your raw CSV header for 'Governorate'.")


# 2. Visualizing data by Districts (Proportions)
if 'district' in df_completed.columns:
    print("\n## 2. Visualizing data by Districts (Proportions)")
    print("Objective: Provide a more granular view of disparities at the district level, crucial for hyper-local intervention planning.")

    # Example 1: Proportion of Households with Disabilities by District
    if 'disabilities' in df_completed.columns:
        plt.figure(figsize=(14, 7))
        disability_by_dist = df_completed.groupby('district')['disabilities'].value_counts(normalize=True).unstack().fillna(0)
        if 'yes' in disability_by_dist.columns: # Check if 'yes' is even a response
            sns.barplot(x=disability_by_dist.index, y=disability_by_dist['yes'], palette='viridis')
            plt.title('Proportion of Households with Disabilities by District')
            plt.xlabel('District')
            plt.ylabel('Proportion of Households with Disabilities (Yes)')
            plt.ylim(0, 1)
            plt.xticks(rotation=90, ha='center', fontsize=8)
            plt.tight_layout()
            plt.show()
            print("Interpretation: This shows disability prevalence at a finer district level. **Caution:** Be mindful if some districts have very few samples, as small sample sizes can lead to unstable estimates and misleading comparisons.")
        else:
            print("No 'yes' disabilities found for detailed district analysis.")
    else:
        print("Warning: 'disabilities' column (mapped from 'Disabilities in Household?') not found for district analysis.")

    # Example 2: Filter by a specific Governorates (e.g., 'Beirut') then analyze districts within it
    if 'governorate' in df_completed.columns and 'no_members_per_day' in df_completed.columns:
        # Filter for the *original* name of the governorate
        # Assuming 'Beirut' is a possible value in the 'Governorate' column (before renaming to 'governorate')
        df_beirut = df_completed[df_completed['governorate'] == 'beirut'].copy() # Use the mapped name 'beirut'
        if not df_beirut.empty:
            avg_meals_beirut_dist = df_beirut.groupby('district')['no_members_per_day'].mean().sort_values(ascending=False)
            plt.figure(figsize=(8, 5))
            sns.barplot(x=avg_meals_beirut_dist.index, y=avg_meals_beirut_dist.values, palette='plasma')
            plt.title('Average Meals Per Day in Beirut Districts')
            plt.xlabel('District')
            plt.ylabel('Average Meals Per Day')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()
            print("Interpretation: A more focused view on districts within a specific governorate, revealing hyper-local patterns in food security within major urban centers and allowing for targeted resource allocation within a smaller area.")
        else:
            print("No completed interviews found for Beirut governorate for district-level analysis.")
    else:
        print("Warning: 'governorate' (mapped) or 'no_members_per_day' (mapped) column not found for focused district analysis.")
else:
    print("\n## 2. Visualizing data by Districts: 'district' column (mapped from 'District') not found.")


# 3. Relation of Age with Gender (Household Members)
# Use the raw column names for dynamic detection
member_age_cols_raw = [col for col in df_completed.columns if col.startswith('Member ') and ' Age' in col]
member_gender_cols_raw = [col for col in df_completed.columns if col.startswith('Member ') and ' Gender' in col]

if member_age_cols_raw and member_gender_cols_raw:
    print("\n## 3. Relation of Age with Gender (Household Members)")
    print("Objective: Understand the demographic structure of household members by exploring the age distribution for males vs. females, crucial for age- and gender-specific program design.")

    member_data = []
    # Determine the maximum index for repeat group members present in the dataframe
    # We assume 'Member 1 Age' up to 'Member 8 Age'
    for i in range(1, 9): # Iterate for Member 1 to Member 8
        age_col_name = f'Member {i} Age'
        gender_col_name = f'Member {i} Gender'
        if age_col_name in df_completed.columns and gender_col_name in df_completed.columns:
            # Stack data for this member number across all rows
            member_subset = df_completed[[age_col_name, gender_col_name]].copy()
            member_subset.columns = ['age', 'sex'] # Standardize names for plotting
            member_data.append(member_subset)

    if member_data:
        df_members = pd.concat(member_data).dropna(subset=['age', 'sex']).copy()

        df_members['age'] = pd.to_numeric(df_members['age'], errors='coerce')
        df_members = df_members.dropna(subset=['age', 'sex'])

        plt.figure(figsize=(10, 6))
        sns.histplot(data=df_members, x='age', hue='sex', kde=True, bins=20, palette='coolwarm')
        plt.title('Age Distribution of Household Members by Gender')
        plt.xlabel('Age (Years)')
        plt.ylabel('Count')
        plt.show()

        plt.figure(figsize=(7, 5))
        sns.boxplot(data=df_members, x='sex', y='age', palette='coolwarm')
        plt.title('Age Distribution by Gender (Box Plot)')
        plt.xlabel('Gender')
        plt.ylabel('Age (Years)')
        plt.show()

        print("Interpretation: The plots show the typical age ranges for males and females within the households. A KDE plot visualizes the density distribution, and a box plot highlights median ages, interquartile ranges, and outliers for each gender, informing demographic-specific program design.")
    else:
        print("No valid household member data available for age-gender analysis after unpivoting. Check if 'Member X Age'/'Member X Gender' columns exist and contain data.")
else:
    print("\n## 3. Relation of Age with Gender: 'Member X Age' or 'Member X Gender' columns not found. Please check your raw CSV headers.")


# 4. Relation of Legal Status of the Head of the Household with the Number of Meals per day
if 'no_members_per_day' in df_completed.columns and 'head_household_legal_status' in df_completed.columns:
    print("\n## 4. Relation of Legal Status of the Head of the Household with the Number of Meals per day")
    print("Objective: Investigate if there's a correlation between the head of household's legal status and the household's food security (number of meals per day), which is critical for targeting interventions and understanding vulnerability.")

    df_analysis_meals_legal = df_completed.dropna(subset=['no_members_per_day', 'head_household_legal_status']).copy()

    if not df_analysis_meals_legal.empty:
        plt.figure(figsize=(10, 6))
        sns.boxplot(data=df_analysis_meals_legal, x='head_household_legal_status', y='no_members_per_day', palette='cividis')
        plt.title('Meals Per Day by Legal Status of Head of Household')
        plt.xlabel('Legal Status of Head of Household')
        plt.ylabel('Number of Meals Per Day')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

        print("\nAverage meals per day by Legal Status:")
        print(df_analysis_meals_legal.groupby('head_household_legal_status')['no_members_per_day'].mean())

        print("Interpretation: This box plot visually compares the median, spread, and outliers of daily meal consumption for households grouped by the legal status of their head. Differences in medians or ranges can indicate disparities in food security, guiding targeted support and policy recommendations.")
    else:
        print("Not enough relevant data to analyze legal status vs. meals per day after filtering.")
else:
    print("\n## 4. Relation of Legal Status with Meals Per Day: 'no_members_per_day' (mapped from 'Number of Meals Per Day') or 'head_household_legal_status' (mapped from 'Head of Household Legal Status') columns not found.")


print("\n--- Analysis Complete ---")
print("Remember to save your Colab notebook (`File > Save a copy in Drive`) to preserve your code, outputs, and narrative!")
print("\nNext Steps:")
print("1. Review all plots and interpretations.")
print("2. Add more specific markdown text based on what *your* data actually shows.")
print("3. Consider further statistical tests if your sample size allows and research questions require.")
print("4. Prepare a final report or presentation based on your findings.")
print("5. **For Looker Studio:** You will likely need to export the 'df_completed' DataFrame (and 'df_members' if created) into new, separate CSVs/Sheets that are pre-processed for Looker Studio's needs (especially the unpivoted members and possibly split multi-selects).")
```
